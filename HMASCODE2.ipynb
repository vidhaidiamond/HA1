Code:2 Loss Functions & Hyperparameter Tuning

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Define true values (y_true) and model predictions (y_pred)
y_true = np.array([[0, 1, 0], [1, 0, 0]])  # One-hot encoded true labels
y_pred1 = np.array([[0.1, 0.8, 0.1], [0.7, 0.2, 0.1]])  # Initial predictions
y_pred2 = np.array([[0.2, 0.7, 0.1], [0.6, 0.3, 0.1]])  # Slightly modified predictions

# Compute Mean Squared Error (MSE)
mse_loss_fn = tf.keras.losses.MeanSquaredError()
mse1 = mse_loss_fn(y_true, y_pred1).numpy()
mse2 = mse_loss_fn(y_true, y_pred2).numpy()

# Compute Categorical Cross-Entropy (CCE)
cce_loss_fn = tf.keras.losses.CategoricalCrossentropy()
cce1 = cce_loss_fn(y_true, y_pred1).numpy()
cce2 = cce_loss_fn(y_true, y_pred2).numpy()

# Print loss values
print(f"MSE Loss for y_pred1: {mse1:.4f}")
print(f"MSE Loss for y_pred2: {mse2:.4f}")
print(f"CCE Loss for y_pred1: {cce1:.4f}")
print(f"CCE Loss for y_pred2: {cce2:.4f}")

# Plot loss function values
labels = ['MSE y_pred1', 'MSE y_pred2', 'CCE y_pred1', 'CCE y_pred2']
loss_values = [mse1, mse2, cce1, cce2]

plt.figure(figsize=(8, 5))
plt.bar(labels, loss_values, color=['blue', 'cyan', 'red', 'orange'])
plt.ylabel("Loss Value")
plt.title("Comparison of MSE and Categorical Cross-Entropy Loss")
plt.show()
```
